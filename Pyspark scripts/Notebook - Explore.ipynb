## Available Big Tables in Unity Catalog

You have access to several large datasets in the `samples` catalog:

### 1. **NYC Taxi Data** - `samples.nyctaxi.trips`
* **~22K rows** - NYC taxi trip records with fare, distance, location data
* Good for: geospatial analysis, time series, revenue analysis

### 2. **TPC-DS Benchmark Data (Scale Factor 1)** - `samples.tpcds_sf1.*`
* **2.8M+ rows** in `store_sales` table
* **1.4M+ rows** in `catalog_sales` table
* **11.7M+ rows** in `inventory` table
* Complete retail dataset with customer, product, sales, returns data
* Good for: sales analysis, customer segmentation, inventory management

### 3. **TPC-DS Benchmark Data (Scale Factor 1000)** - `samples.tpcds_sf1000.*`
* **1000x larger** than the sf1 tables (billions of rows)
* Same schema as tpcds_sf1 but scaled up
* Good for: big data testing, performance benchmarking

### Other Available Schemas:
* `samples.accuweather` - Weather data
* `samples.healthverity` - Healthcare data
* `samples.sec` - SEC filings data
* `samples.tpch` - TPC-H benchmark data
* `samples.wanderbricks` - Travel/tourism data
